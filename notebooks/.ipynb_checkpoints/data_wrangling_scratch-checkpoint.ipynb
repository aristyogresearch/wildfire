Bin California into latitude, longitude points. Using externaly generated list of points, discard those which are not within California's boundries.

```python
from multiprocessing import Pool

bins = pd.read_csv('spatial_data/bins.csv')
empty = pd.Series([np.nan, np.nan ])
empty.index = ['lat', 'long']

def point_in_california(point):
    """ Takes point and returns point if in california"""
    coord = Point(point['long'], point['lat'])
    if coord.within(california) == True:
        return point
    else:
        return empty
    
def cull_points(points):
    keepers = points.apply(point_in_california, axis=1)
    return keepers
    
def parallelize(df, func, n_cores=14):
    df_split = np.array_split(df, n_cores)
    pool = Pool(n_cores)
    df = pd.concat(pool.map(func, df_split))
    pool.close()
    pool.join()
    return df

keeper_bins = parallelize(bins, cull_points).dropna()
keeper_bins.to_csv('spatial_data/california_bins.csv', index = None, header=True)
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (8,8)
plt.scatter(x=keeper_bins['long'], y=keeper_bins['lat'], color='black', s=0.5, label='bins')
plt.axis('equal')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.title('Latitude, longitude bins')
plt.legend()
plt.show()
```

Curate list of CDEC weather monitoring stations, discard those which are not actualy in california.

```python
stations = pd.read_csv('weather_station_data/target_stations.csv')
empty = pd.Series([np.nan, np.nan ])
empty.index = ['latitude', 'longitude']

def point_in_california(point):
    """ Takes point and returns point if in california"""
    coord = Point(point['longitude'], point['latitude'])
    if coord.within(california) == True:
        return point
    else:
        return empty
    
def cull_points(points):
    keepers = points.apply(point_in_california, axis=1)
    return keepers
    
def parallelize(df, func, n_cores=14):
    df_split = np.array_split(df, n_cores)
    pool = Pool(n_cores)
    df = pd.concat(pool.map(func, df_split), sort=False)
    pool.close()
    pool.join()
    return df

keeper_stations = parallelize(stations, cull_points).dropna()
keeper_stations.to_csv('weather_station_data/target_stations_in_california.csv', index = None, header=True)
```

```python
plt.rcParams["figure.figsize"] = (8,8)
plt.scatter(x=keeper_bins['long'], y=keeper_bins['lat'], color='black', s=0.5, label='bins')
plt.scatter(x=keeper_stations['longitude'], y=keeper_stations['latitude'], color='green', s=2, label='bins')
plt.axis('equal')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.title('Weather station locations')
plt.legend()
plt.show()
```

```python
fires = pd.read_csv('training_data/wild_fire_data/FPA_FOD_20170508.csv', low_memory=False)
fire_locations = fires[['LATITUDE', 'LONGITUDE']].copy()
fire_locations.head()
```

Since we have many bins which do not contain a weather monitoring station, the next step is to fill in their values from the nearest station. First find the nearest station to each bin.

```python
keeper_bins = pd.read_csv('spatial_data/california_bins.csv')
bin_array = np.column_stack([keeper_bins['long'], keeper_bins['lat']])
bin_array
```

```python
keeper_stations = pd.read_csv('weather_station_data/target_stations_in_california.csv')
station_array = np.column_stack([keeper_stations['longitude'], keeper_stations['latitude']])
station_array
```

```python
from scipy import spatial

station_tree = spatial.cKDTree(station_array)
dist, indexes = station_tree.query(bin_array)
```

```python
nearest_station_names = []
for index in indexes:
    nearest_station_names.append(keeper_stations.iloc[index, 3])
```

```python
keeper_bins['nearest_station_name'] = nearest_station_names
keeper_bins.head()
```
